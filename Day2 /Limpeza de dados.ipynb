### Criando uma coluna condicional:



```python
def classificacao_CDU(localizacao):
    if localizacao >= 0 and localizacao <= 99:
        return 'Generalidades. Ci√™ncia e conhecimento'
    elif localizacao >= 100 and localizacao <= 199:
        return 'Filosofia e psicologia'
    elif localizacao >= 200 and localizacao <= 299:
        return 'Religi√£o'
    elif localizacao >= 300 and localizacao <= 399:
        return 'Ci√™ncias Sociais'
    elif localizacao >= 400 and localizacao <= 499:
        return 'Classe vaga. Provisoriamente n√£o ocupada'
    elif localizacao >= 500 and localizacao <= 599:
        return 'Matem√°tica e ci√™ncias naturais'
    elif localizacao >= 600 and localizacao <= 699:
        return 'Ci√™ncias aplicadas'
    elif localizacao >= 700 and localizacao <= 799:
        return 'Belas artes'
    elif localizacao >= 800 and localizacao <= 899:
        return 'Linguagem. L√≠ngua. Lingu√≠stica'
    elif localizacao >= 900 and localizacao <= 999:
        return 'Geografia. Biografia. Hist√≥ria'
    else:
        return 'Classifica√ß√£o desconhecida'
```
